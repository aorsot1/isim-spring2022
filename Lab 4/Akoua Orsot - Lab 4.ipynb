```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
cap_data = pd.read_csv("lab4_calibration_data.csv")
```

```python
# Regression between Angle and Voltage 
x = cap_data["Capacitance (pF)"].values
y = cap_data["Voltage (V)"].values
X, Y = x.reshape(-1,1), y.reshape(-1,1)
reg = LinearRegression().fit(X, Y)
y_pred = reg.predict(X) 
intercept = round(reg.intercept_[0], 5)
slope = round(reg.coef_[0][0], 5)
reg_eq =  'V = {:.5f} + {:.5f}C'.format(intercept, slope)
reg_eq
```

```python
#Plotting temp vs. time
fig = plt.figure(figsize=(10, 6), dpi=80)

plt.plot(X, y_pred, 'r--', label=reg_eq)
plt.scatter(x, y)
plt.title("6 different capacitors were used to monitor RMS voltage Amplitude", 
             y=1.05, fontsize=10)
plt.suptitle("Circuit Calibration - Capacitance vs. Voltage")
plt.xlabel("Capacitance (pF)")
plt.ylabel("Voltage (V)")

plt.show()
fig.savefig("Circuit Calibration.png")
```
