```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
hot_water = pd.read_csv("hot_water.csv")
hot_water_ice = pd.read_csv("hot_water_ice.csv")
```

```python
time = hot_water.iloc[0,0]
time_lapse_0 = hot_water['t1'] - time
voltage_0 = hot_water['ch1']

time_1 = hot_water_ice.iloc[0,0]
time_lapse_1 = hot_water_ice['t1'] - time_1
voltage_1 = hot_water_ice['ch1']
```

```python
fig = plt.figure(figsize=(10, 6), dpi=80)

plt.plot(time_lapse_0, voltage_0 )
plt.plot(time_lapse_1, voltage_1)
plt.suptitle("Voltage vs. Time")
plt.xlabel("Time (Seconds)")
plt.ylabel("Voltage (Volts)")

plt.show()

#fig.savefig("Calibration Curve.png")
```
