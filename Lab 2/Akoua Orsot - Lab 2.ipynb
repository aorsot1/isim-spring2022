```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
hot_water = pd.read_csv("hot_water.csv")
hot_water_ice = pd.read_csv("hot_water_ice.csv")
```

```python
## Calculating change in time and voltage

# Experiment 1: Hot Water alone
time_delta0 = hot_water['t1'] - hot_water.iloc[0,0]
voltage_delta0 = hot_water['ch1'] 
#- hot_water.iloc[0,1]

# Experiment 1: Hot Water + 20mL Ice Water
time_delta1 = hot_water_ice['t1'] - hot_water_ice.iloc[0,0]
voltage_delta1 = hot_water_ice['ch1'] 
#- hot_water_ice.iloc[0,1]
```

```python
fig = plt.figure(figsize=(10, 6), dpi=80)

plt.plot(time_delta0, voltage_delta0)
plt.plot(time_delta1, voltage_delta1)
plt.suptitle("Voltage vs. Time")
plt.xlabel("Time (Seconds)")
plt.xlim([0, 180])
plt.ylabel("Voltage (Volts)")

plt.show()

#fig.savefig("Calibration Curve.png")
```

```python
# Deriving the Temperature from the formula
voltage = hot_water['ch1']
temp = 3528 / (np.log(voltage/1000) * (-3528/298))
voltage1 = hot_water_ice['ch1']
temp1 = 3528 / (np.log(voltage1/1000) * (-3528/298))
```

```python
#Plotting temp vs. time
fig = plt.figure(figsize=(10, 6), dpi=80)

plt.plot(time_delta0, temp)
plt.plot(time_delta1, temp1)
plt.suptitle("Temperature vs. Time")
plt.xlabel("Time (Seconds)")
plt.xlim([0, 180])
plt.ylabel("Temperature (Kelvin)")

plt.show()
```
